```python
player_data = pd.read_json("Data/kaggle_player_data.json")
```

```python
player_data.head()
```

```python
player_data.to_csv("Data/kaggle_player_data.csv")
```

```python
player_data["position"].unique()
```

```python
position_data = player_data["position"]
```

```python
position_cleanup = []
for x in position_data:
    position_cleanup.append(x[:2])
```

```python
player_data["position_cleaned"] = position_cleanup
```

```python
player_data.head()
```

```python
player_data["position_cleaned"].unique()
```

```python
offensive_positions =['QB','RB','TB','HB','WR','TE','FB','WB','SE','FL']
offensive_positions
```

```python
offensive_position_players = player_data[player_data["position_cleaned"].isin(offensive_positions)]
```

```python
offensive_position_players.head()
```

```python
offensive_position_players.loc[offensive_position_players["position_cleaned"] == "TB", "position_cleaned"] = "RB"
```

```python
offensive_position_players.loc[offensive_position_players["position_cleaned"] == "HB", "position_cleaned"] = "RB"
```

```python
offensive_position_players.loc[offensive_position_players["position_cleaned"] == "FB", "position_cleaned"] = "RB"
```

```python
offensive_position_players.loc[offensive_position_players["position_cleaned"] == "WB", "position_cleaned"] = "RB"
```

```python
offensive_position_players.loc[offensive_position_players["position_cleaned"] == "SE", "position_cleaned"] = "WR"
```

```python
offensive_position_players.loc[offensive_position_players["position_cleaned"] == "FL", "position_cleaned"] = "WR"
```

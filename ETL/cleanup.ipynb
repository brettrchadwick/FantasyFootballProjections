```python
import os
import pandas as pd
```

```python
#Resource Files
games='../Resources/games.json'
profiles='../Resources/profiles.json'
```

```python
#read into DataFrame
games_df=pd.read_json(games)
profiles_df=pd.read_json(profiles)
```

```python
profiles_df.head()
```

```python
games_df.head()
```

```python
#Cleaning up data for players listed with multiple positions, the primary one is the one kept.
position_data = profiles_df["position"]
position_cleanup = []
for x in position_data:
    position_cleanup.append(x[:2])
profiles_df["position_cleaned"] = position_cleanup
```

```python
#separating offense and defense
offensive_positions =['QB','RB','TB','HB','WR','TE','FB','WB','SE','FL']
offensive_position_players = profiles_df[profiles_df["position_cleaned"].isin(offensive_positions)]
offensive_position_players.head()
```

```python
#combining different named positions to RB
offensive_position_players.loc[offensive_position_players["position_cleaned"] == "TB", "position_cleaned"] = "RB"
offensive_position_players.loc[offensive_position_players["position_cleaned"] == "HB", "position_cleaned"] = "RB"
offensive_position_players.loc[offensive_position_players["position_cleaned"] == "FB", "position_cleaned"] = "RB"
offensive_position_players.loc[offensive_position_players["position_cleaned"] == "WB", "position_cleaned"] = "RB"
offensive_position_players.loc[offensive_position_players["position_cleaned"] == "SE", "position_cleaned"] = "WR"
offensive_position_players.loc[offensive_position_players["position_cleaned"] == "FL", "position_cleaned"] = "WR"
offensive_position_players.to_csv('../Resources/offensive_players.csv')
```

```python
#filtering all offensive players into one DataFrame, and only pulling necessary columns
all_offensive_players=offensive_position_players['player_id']
total_offensive_stats = games_df[games_df["player_id"].isin(all_offensive_players)]
total_offensive_stats=total_offensive_stats.iloc[:,0:32]
total_offensive_stats.to_csv('../Resources/total_offensive_stats.csv')
```

```python
total_offensive_stats.nunique()
```

```python
#calculate total passing fantasy points
total_offensive_stats["passing_yards_points"] = total_offensive_stats["passing_yards"].divide(25)
total_offensive_stats["passing_touchdowns_points"] = total_offensive_stats["passing_touchdowns"].multiply(6)
total_offensive_stats["passing_interceptions_points"] = total_offensive_stats["passing_interceptions"].multiply(-2)
total_offensive_stats["passing_sacks_points"] = total_offensive_stats["passing_sacks"].multiply(-1)
total_offensive_stats["total_passing_fantasy_points"] = total_offensive_stats["passing_yards_points"] + total_offensive_stats["passing_touchdowns_points"] + total_offensive_stats["passing_interceptions_points"] + total_offensive_stats["passing_sacks_points"]
total_offensive_stats = total_offensive_stats.drop(columns=["passing_yards_points","passing_touchdowns_points","passing_interceptions_points","passing_sacks_points"])
```

```python
#calculate total rushing fantasy points
total_offensive_stats["rushing_yards_points"] = total_offensive_stats["rushing_yards"].divide(10)
total_offensive_stats["rushing_touchdowns_points"] = total_offensive_stats["rushing_touchdowns"].multiply(6)
total_offensive_stats["total_rushing_fantasy_points"] = total_offensive_stats["rushing_yards_points"] + total_offensive_stats["rushing_touchdowns_points"]
total_offensive_stats = total_offensive_stats.drop(columns=["rushing_yards_points","rushing_touchdowns_points"])
```

```python
#calculate total receiving fantasy points
total_offensive_stats["receiving_receptions_points"] = total_offensive_stats["receiving_receptions"].multiply(1)
total_offensive_stats["receiving_yards_points"] = total_offensive_stats["receiving_yards"].divide(10)
total_offensive_stats["receiving_touchdowns_points"] = total_offensive_stats["receiving_touchdowns"].multiply(6)
total_offensive_stats["total_receiving_fantasy_points"] = total_offensive_stats["receiving_receptions_points"] + total_offensive_stats["receiving_yards_points"] + total_offensive_stats["receiving_touchdowns_points"]
total_offensive_stats = total_offensive_stats.drop(columns=["receiving_receptions_points","receiving_yards_points","receiving_touchdowns_points"])
```

```python
#calculate total other fantasy points
total_offensive_stats["total_other_fantasy_points"] = total_offensive_stats["kick_return_touchdowns"].multiply(6) + total_offensive_stats["punt_return_touchdowns"].multiply(6)
```

```python
total_offensive_stats["total_combined_fantasy_points"] = total_offensive_stats["total_passing_fantasy_points"] + total_offensive_stats["total_rushing_fantasy_points"] + total_offensive_stats["total_receiving_fantasy_points"] + total_offensive_stats["total_other_fantasy_points"]  
```

```python
total_offensive_stats.to_csv('../Resources/total_offensive_stats.csv')
```
